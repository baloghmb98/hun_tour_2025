```python
import folium

# Define the map centered at a location
hun_admin2 = folium.Map(location=[47.0, 19.0], zoom_start=5)


# URL of your GeoJSON file hosted on GitHub
geojson_url = "https://raw.githubusercontent.com/wuerdo/geoHungary/refs/heads/master/counties.geojson"


# Add the GeoJSON layer
folium.GeoJson(geojson_url, name="geojson").add_to(hun_admin2)


# Add layer control
folium.LayerControl().add_to(hun_admin2)


#####folium.Map(tiles="cartodbpositron")

# Save the map to an HTML file
hun_admin2.save("hun_admin2_map.html")
```
